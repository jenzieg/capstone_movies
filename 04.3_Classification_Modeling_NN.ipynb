# Movie Classification Modeling - Neural Networks 

### Import Libraries & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras import utils, models
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score


from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# importing warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
# https://machinelearningmastery.com/how-to-fix-futurewarning-messages-in-scikit-learn/

pd.set_option("display.max_columns",350)
```

```python
df = pd.read_csv('/content/df_model.csv')
df.drop(columns = 'Unnamed: 0', inplace = True)
print(df.shape)
df.head()
```

### Setup for Modeling

```python
X = df.drop(columns = ['audience_rating', 'audience_status', 'audience_count'])
y = df['audience_status']
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, test_size=0.30, random_state=321)
```

```python
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

```python
y_test.value_counts(normalize=True)
```

### Neural Network Binary Classification

```python
model = Sequential()

# First Dense Layer
model.add(Dense(16,
               activation = 'relu',  
               input_shape = Z_train[0].shape)) 

# Second hidden layer
model.add(Dense(16, activation = 'relu'))

# Output layer 
model.add(Dense(1, activation = 'sigmoid')) 
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy','Recall','Precision'])
```

```python
model = Sequential()

# Add input and hidden layer
model.add(Dense(32,
               activation = 'relu',
               input_shape = Z_train[0].shape))
model.add(Dropout(0.05))

# Add 2nd & 3rd hidden layer
model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.05))

model.add(Dense(32, activation = 'relu'))
model.add(Dropout(0.05))

# Output layer
model.add(Dense(1, activation = 'sigmoid'))

# Adding in Early Stopping
es = EarlyStopping(patience=5)

# Compile our data
model.compile(optimizer = 'adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```
